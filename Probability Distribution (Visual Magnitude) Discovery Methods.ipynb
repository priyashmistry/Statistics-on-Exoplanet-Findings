# Plot the graph 

```python
x = np.linspace(0, np.size(xlabel)-1, np.size(xlabel))
X_ = np.arange(x.min(), x.max(), 0.1)

X_Y_Splinet = make_interp_spline(x, countt)
X_Y_Splinets = make_interp_spline(x, countts)
X_Y_Splinerv = make_interp_spline(x, countrv)
X_Y_Splinedi = make_interp_spline(x, countdi)
X_Y_Splinem = make_interp_spline(x, countm)
X_Y_Splineo = make_interp_spline(x, counto)

Y_t = X_Y_Splinet(X_)
Y_ts = X_Y_Splinets(X_)
Y_rv = X_Y_Splinerv(X_)
Y_di = X_Y_Splinedi(X_)
Y_m = X_Y_Splinem(X_)
Y_o = X_Y_Splineo(X_)

plt.fill_between(X_,Y_t, zorder = 3, color='crimson', alpha = 0.6, label = "Transit Method (Ground Based)")

plt.fill_between(X_,Y_ts, zorder = 4, color='b', alpha = 0.6, label = "Transit Method (Space Based)")

plt.fill_between(X_,Y_rv, zorder = 2, color='springgreen', alpha = 0.6, label = "Radial Velocity Method")

plt.fill_between(X_,Y_di, zorder = 5, color='deepskyblue', alpha = 0.6, label = "Direct Imaging")

plt.fill_between(X_,Y_m, zorder = 6, color='black', alpha = 0.6, label = "Gravitational Microlensing")

plt.fill_between(X_,Y_o, zorder = 7, color='yellow', alpha = 0.6, label = "Other Methods")

plt.xlabel('V Magnitude',fontdict=font1, weight = "bold",fontsize="20")
plt.ylabel('Number of Exoplanets',fontdict=font1, weight = "bold",fontsize="20")

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.grid(axis = "y", zorder = 1)

#plt.ylim(0,15)

plt.legend(fontsize=15)

fig=plt.gcf()
fig.set_size_inches(15,10)

#fig.savefig(os.path.join('Other_Facilities_Methods.jpeg'), format='jpeg', dpi=600, bbox_inches = 'tight')
```
