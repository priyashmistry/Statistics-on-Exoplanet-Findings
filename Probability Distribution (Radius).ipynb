# Calculate the Occurrecne Rate 

```python
count = np.array([c1, c2, c3, c4, c5, c6, c7, c8, c9, c10, c11, c12, c13, c14, c15, c16, c17, c18, c19, c20, c21, c22, c23, c24, c25, c26, c27, c28, c29, c30, c31, c32, c33, c34])

if sum(count) == np.size(radius):
    print("True")
    print("Sum = ",sum(count), "\nNo. of Rows = ",np.size(radius))
else:
    print("False")
    print("Sum = ",sum(count), "\nNo. of Rows = ",np.size(radius))
    
ocrate = count*100/np.size(radius)
```

# Find out the % of host stars having radii range defined above

```python
counter_array = np.zeros([34,10], dtype = int) # Array to store the calculated numbers

# Loop that calculates the number of exoplanets in the respective spectral type of host star, 
# which falls into the given radius range
for n in range(np.size(sradius_values)):
    for m in range(np.size(sradius_values[n])):
        if sradius_values[n][m] > 0 and sradius_values[n][m] <= 0.5:
            counter_array[n][0] = counter_array[n][0] + 1 
                    
        elif sradius_values[n][m] > 0.5 and sradius_values[n][m] <= 1.0:
            counter_array[n][1] = counter_array[n][1] + 1
        
        elif sradius_values[n][m] > 1.0 and sradius_values[n][m] <= 1.5:
            counter_array[n][2] = counter_array[n][2] + 1
                    
        elif sradius_values[n][m] > 1.5 and sradius_values[n][m] <= 2.0:
            counter_array[n][3] = counter_array[n][3] + 1
                
        elif sradius_values[n][m] > 2.0 and sradius_values[n][m] <= 2.5:
            counter_array[n][4] = counter_array[n][4] + 1
        
        elif sradius_values[n][m] > 2.5 and sradius_values[n][m] <= 3.0:
            counter_array[n][5] = counter_array[n][5] + 1
        
        elif sradius_values[n][m] > 3.0 and sradius_values[n][m] <= 4.0:
            counter_array[n][6] = counter_array[n][6] + 1
                    
        elif sradius_values[n][m] > 4.0 and sradius_values[n][m] <= 6.0:
            counter_array[n][7] = counter_array[n][7] + 1
                
        elif sradius_values[n][m] > 6.0 and sradius_values[n][m] <= 10.0:
            counter_array[n][8] = counter_array[n][8] + 1
        
        elif sradius_values[n][m] > 10:
            counter_array[n][9] = counter_array[n][9] + 1

rad_data = np.transpose(counter_array)
```

# Calculate the Occerrence Rate of Host Stars' radii

```python
ocrate_n = ["ocrate1", "ocrate2", "ocrate3", "ocrate4", "ocrate5", "ocrate6", "ocrate7", "ocrate8", "ocrate9", "ocrate10", \
            "ocrate11", "ocrate12", "ocrate13", "ocrate14", "ocrate15", "ocrate16", "ocrate17", "ocrate18", "ocrate19", "ocrate20", \
            "ocrate21", "ocrate22", "ocrate23", "ocrate24", "ocrate25", "ocrate26", "ocrate27", "ocrate28", "ocrate29", "ocrate30", \
            "ocrate31", "ocrate32", "ocrate33", "ocrate34"]

count_n = ["count1", "count2", "count3", "count4", "count5", "count6", "count7", "count8", "count9", "count10", \
           "count11", "count12", "count13", "count14", "count15", "count16", "count17", "count18", "count19", "count20", \
           "count21", "count22", "count23", "count24", "count25", "count26", "count27", "count28", "count29", "count30", \
           "count31", "count32", "count33", "count34"]

for i in range(np.size(ocrate_n)):
    if sum(counter_array[i]) == 0:
        ocrate_n[i] = np.zeros(10)
    else:
        count_n[i] = np.array(counter_array[i])
        ocrate_n[i] = count_n[i]*ocrate[i]/sum(count_n[i])
```

# Graph 1 : Simple Bar Graph showing the Statistics of Exoplanets' Radii

```python
y_dashed = np.linspace(0, int(max(ocrate)) + 1 , 100)
x_const = np.ones(np.size(y_dashed))

x = np.linspace(0, np.size(xlabel)-1, np.size(xlabel))
X_Y_Spline = make_interp_spline(x, ocrate)

X_ = np.arange(x.min(), x.max(), 0.1)
Y_ = X_Y_Spline(X_)

plt.bar(xlabel,ocrate, zorder = 2, color="lightsalmon", edgecolor='orangered', linewidth = 1.5, width = 0.6)
plt.plot(X_, Y_, 'g', zorder = 3, linewidth = 2)
plt.plot(x_const, y_dashed, '--k')
plt.plot(x_const*11, y_dashed, '--k')
plt.plot(x_const*3, y_dashed, '--k')

plt.title("Radius : Probability Distribution", fontdict=font1, weight = "bold",fontsize="30")
plt.xlabel('Radius ($R_E$)',fontdict=font1, weight = "bold",fontsize="20")
plt.ylabel('Occurrence Rate (%)',fontdict=font1, weight = "bold",fontsize="20")

plt.xticks(fontsize = 15, rotation = 'vertical')
plt.yticks(fontsize = 15)
plt.grid(axis = "y", zorder = 1)

props = dict(boxstyle='round', facecolor='lightcyan')

plt.text(21, 4, "Total No. of Exoplanets : %d"%(np.size(radius)), fontsize=20, verticalalignment='top', bbox=props)
plt.text(-0.3, 12.7, "Earth", rotation = 90, fontsize="15")
plt.text(3.5, 12.0, "Neptune", rotation = 90, fontsize="15")
plt.text(10, 12.3, "Jupiter", rotation = 90, fontsize="15")

fig=plt.gcf()
fig.set_size_inches(15,7)
```
